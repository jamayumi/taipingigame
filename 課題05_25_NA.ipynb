# **タイピングゲーム**

## **概要**
このゲームはタイピングを練習することができるタイピングゲームとなっています。
スコアを最後に表示させることで、スコアを見てタイピングスキルが上がっているかがわかりやすくなっています。



## **操作方法**
まず、【サブ処理】の**Player_name** を入力して頂き、**mode**を設定してください。

**mode** によって出てくるお題が変わってきます。

現在用意している**mode**は【Web3用語】【普通】【動物の名前】【英単語】【四字熟語】です。

その後「メイン処理」を実行してください。

実行すると10秒後にタイピングゲームが始まります。

お題が表示されるのでユーザー入力欄に入力してください。入力が終わったらEnterキーを押して次のお題に行ってください。


## **制作者のコメント**

タイピングはそれ程上手くないですが、タイピング自体がとても好きなので今回タイピングゲームを作りました。思ったように動かないことが多かったので修正に時間がかかってしまいました。また、modeを選択してそのmodeにあったお題を出すのが難しかったです。作っていてとても勉強になりましたし、楽しかったです。

```python
#@title サブ処理
import random
import time
from IPython.core.display import clear_output

def get_level(average_score):
    if average_score >= 200:
        return "S\nプロレベルです。非常に高いタイピングスキルを持っています！さらなる向上を目指しましょう。"
    elif average_score >= 150:
        return "A+\n一般的なオフィスワークでは困ることのないレベルです。プロレベルにはまだ届きません。より高いレベルを目指すには、正確さを高めましょう。"
    elif average_score >= 100:
        return "A\n一般的なオフィスワークでは困ることのないレベルです。プロレベルにはまだ届きません。より高いレベルを目指すには、正確さを高めましょう。"
    elif average_score >= 80:
        return "B\nキーの配置とタイピングの基本操作が理解できてきました。できるだけ毎日練習を続けましょう。"
    elif average_score >= 70:
        return "C\nタイピングが生まれつきできる人はいません。誰もが努力してできるようになっていきます。ここからスタートです。"
    elif average_score >= 60:
        return "D\nタイピングが生まれつきできる人はいません。誰もが努力してできるようになっていきます。ここからスタートです。"
    else:
        return "E\nタイピングが生まれつきできる人はいません。誰もが努力してできるようになっていきます。ここからスタートです。"

def typing_game():
    #@markdown ### タイピングゲーム初期設定
    Player_name = 'くろねこ' #@param {type:'string'}
    mode = '\u56DB\u5B57\u719F\u8A9E' #@param ['Web3用語', '普通','動物の名前','英単語','四字熟語']

    if mode == "Web3用語":
        target_string = [
    "ブロック",
    "チェーン",
    "メタマスク",
    "トークン",
    "ウォレット",
    "コントラクト",
    "データ",
    "ノード",
    "ネットワーク",
    "スマート",
    "メタ",
    "リンク",
    "ダップ",
    "ガバナンス",
    "アドレス",
    "トランザクション",
    "マイニング",
    "ガス",
    "リカバリー",
    "プライバシー",
    "インターフェース",
    "インセンティブ",
    "パブリック",
    "イーサ",
    "ネム",
    "ポリゴン",
]
    elif mode == "普通":
        target_string = [
    "挑戦することで、新たな可能性が広がります。",
    "成長は、挑戦を受け入れた時に始まります。",
    "人生の最大の成果は、困難に立ち向かったときに得られます。",
    "恐れを乗り越える勇気は、自己成長の第一歩です。",
    "限界を押し上げる挑戦は、人をより強くします。",
    "失敗は挫折ではなく、次なる挑戦への準備です。",
    "成長するためには、自分に厳しく、そして寛容であることが必要です。",
    "成功の裏には、果敢に挑戦した姿勢があります。",
    "挑戦は、輝かしい未来への扉を開く鍵です。",
    "たとえ失敗しても、挑戦する勇気を持つことが大切です。",
    "チャレンジは、自分自身の可能性を知る素晴らしい方法です。",
    "成長は、忍耐と積極的な挑戦の結果です。",
    "自分の限界を信じないで、新たな高みに挑戦しましょう。",
    "大胆な挑戦こそが、素晴らしい未来を築く鍵です。",
    "人生の冒険は、挑戦と発見の連続です。",
    "成長は快適な領域を離れて初めて生じます。",
    "自分に挑戦することで、自分自身をより深く理解できます。",
    "失敗を恐れずに挑戦することで、成功への道が開けます。",
    "勇気を持って挑戦すれば、達成感があなたを包み込みます。",
    "一歩踏み出す勇気が、大きな成長をもたらします。"
]

    elif mode == "動物の名前":
        target_string = [
    "犬",
    "猫",
    "象",
    "ライオン",
    "カンガルー",
    "キリン",
    "パンダ",
    "クマ",
    "ウサギ",
    "ペンギン",
    "サル",
    "ワニ",
    "オウム",
    "タコ",
    "ハチ",
    "イルカ",
    "ハト",
    "ゾウリムシ",
    "フクロウ",
    "サメ"
]
    elif mode == "英単語":
        target_string = [
    "cat",
    "dog",
    "book",
    "apple",
    "tree",
    "sun",
    "happy",
    "friend",
    "water",
    "bird",
    "jump",
    "run",
    "ball",
    "flower",
    "blue",
    "smile",
    "school",
    "fish",
    "star",
    "home"
]
    elif mode == "四字熟語":
        target_string = [
    "自業自得",
    "心機一転",
    "虚心坦懐",
    "自縄自縛",
    "無我夢中",
    "自画自賛",
    "不易流行",
    "百花繚乱",
    "無知蒙昧",
    "明鏡止水",
    "自暴自棄",
    "一期一会",
    "三位一体",
    "自家製作",
    "半信半疑",
    "先手必勝"
]



    total_time = 0
    total_score = 0








    # 10秒間のカウントダウン
    for i in range(10, 0, -1):
        print(f"ゲーム開始まで{i}秒...")
        time.sleep(1)
        clear_output(wait=True)

    print(f"{Player_name}さんはMODE:{mode}を選択しました。タイピングゲームを開始します")


    for i in range(10):  # 10回の試行に変更
        random_text = random.choice(target_string)
        print(f"お題({i + 1}/10): {random_text}")

        start_time = time.time()

        user_input = input("ユーザー入力欄: ")

        end_time = time.time()
        elapsed_time = end_time - start_time

        if user_input == random_text:
            print("正解！")
            # 基準時間を10秒とする
            reference_time = 10.0
            # スコアを計算
            score = reference_time / elapsed_time * 100
            level = get_level(score)
            print(f"経過時間: {elapsed_time:.2f}秒")
            print(f"タイピングスコア: {score:.2f}")
            print(f"レベル: {level}")
            total_time += elapsed_time
            total_score += score
        else:
            print("不正解...")
            penalty_score = -50
            total_score += penalty_score

        # ゲームごとにクリア
        clear_output(wait=True)

    average_score = total_score / 10


    print(f"{Player_name}さんの今回のタイピング結果")
    print(f"スコア: {average_score:.2f}")
    print(f"レベル: {level}")
    print(f"入力時間: {total_time:.2f}秒")
```
